# **Project Summary -**

Write the summary here within 500-600 words.
Artificial Intelligence (AI), especially Generative AI tools such as ChatGPT, GitHub Copilot, and Gemini, has transformed the way students learn, complete assignments, and prepare for examinations. While these technologies offer significant advantages in improving productivity and simplifying complex concepts, excessive dependence on AI may negatively impact learning, creativity, knowledge retention, and mental well-being. Understanding the balance between productive AI usage and overdependence has become an important concern for educators and institutions.

This project performs an Exploratory Data Analysis (EDA) on a comprehensive dataset containing information from 50,000 students across different academic disciplines and study levels. The dataset includes variables related to academic performance, AI usage behavior, study habits, prompt engineering skills, institutional AI policies, anxiety levels, burnout risk, and skill retention. The analysis aims to understand how AI usage influences student outcomes and whether responsible AI adoption contributes positively to education.

The project begins with understanding the structure and quality of the dataset through exploratory techniques such as checking data types, duplicate values, missing values, descriptive statistics, and unique categories. After ensuring that the data is clean and suitable for analysis, various visualizations are created to identify important trends and relationships.

Univariate analysis is performed to understand the distribution of individual variables such as weekly AI usage, burnout levels, study hours, academic majors, and prompt engineering skills. Bivariate analysis investigates relationships such as AI usage versus GPA, AI dependency versus burnout, prompt engineering skill versus academic performance, and institutional policy versus skill retention. Multivariate visualizations including correlation heatmaps and pair plots help uncover complex interactions among multiple variables.

The insights obtained from this analysis can help educational institutions design balanced AI policies, encourage healthy AI usage habits, and identify students at higher risk of burnout. The findings may also guide educators in promoting effective prompt engineering skills while maintaining traditional learning methods. Ultimately, this project demonstrates how data-driven decision-making can maximize the educational benefits of AI while minimizing its potential drawbacks.
