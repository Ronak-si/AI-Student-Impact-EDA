## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.
# Create a copy of dataset

df1 = df.copy()
````

````python
df1.dtypes
````

````python
df1["Paid_Subscription"] = df1["Paid_Subscription"].astype("category")
````

````python
# Convert object columns into category

cat_columns = df1.select_dtypes(include='object').columns

for col in cat_columns:
    df1[col] = df1[col].astype("category")
````

````python
# Verify changes

df1.info()
````

````python
# Check Outliers using Boxplots

num_columns = df1.select_dtypes(include=['int64','float64']).columns

plt.figure(figsize=(18,12))

for i,col in enumerate(num_columns):
    plt.subplot(3,3,i+1)
    sns.boxplot(y=df1[col])

plt.tight_layout()
plt.show()
````

````python
# Correlation among numerical columns

df1.corr(numeric_only=True)
````
